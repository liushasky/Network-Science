```python
import random
%matplotlib inline
import networkx as nx
import community as community_louvain
import matplotlib.pyplot as plt
from networkx.algorithms import community
import itertools
from sklearn.metrics.cluster import normalized_mutual_info_score
import statistics
```

## Q1. Write a program that, given a value of the expected external degree $⟨k^{ext}⟩$ in input, generates a GN benchmark. Follow these steps:
* Assign labels to nodes in the same group. So, nodes 0 through 31 have label c1, nodes 32 through 63 have c2, nodes 64 through 95 have c3, and nodes 96 through 127 have c4.  
* Calculate the probabilities pext = $⟨k^{ext}⟩$/96 and pint = (16 − $⟨k^{ext}⟩$)/31.  
* Loop over all pairs of nodes. If two nodes have the same label, i.e., they are in the same group, add a link with probability pint, otherwise with probability pext.


```python
# define a GN benchmark function
def GN_benchmark(kext):
    
    # generate empty network
    G = nx.Graph()
    
    # add nodes in the network
    N = 128
    G.add_nodes_from(range(N))
    
    # assign label to each node in each group
    for n in G.nodes:
        if n < 32:
            d1 = {}
            d1[n] = "c1"
            nx.set_node_attributes(G, d1, "labels")
        elif 31 < n < 64:
            d2 = {}
            d2[n] = "c2"
            nx.set_node_attributes(G, d2, "labels")
        elif 63 < n < 96:
            d3 = {}
            d3[n] = "c3"
            nx.set_node_attributes(G, d3, "labels")
        elif n > 95:
            d4 = {}
            d4[n] = "c4"
            nx.set_node_attributes(G, d4, "labels")
    
    # number of groups
    q = 4
    
    # nodes per group
    nc = 32
    
    # get p_ext and p_int
    p_ext = kext/96
    p_int = (16-kext)/31
    
    # for any combination of node pairs
    for i, j in itertools.combinations(G.nodes, 2):
        r = random.random()
        
        # first check if the random value is larger than the internal link probability
        if G.nodes[i]["labels"] == G.nodes[j]["labels"]:
            if r < p_int:
                G.add_edge(i, j)
        else:
            if r < p_ext:
                G.add_edge(i, j)
                
    return G, q, nc, p_int, p_ext
```

## Q2. Louvain algorithm & Label propagation (asyn)
Test the two community algorithms on the GN benchmark (Q1).   
Use the following values for the expected external degree of the GN benchmark: $⟨k^{ext}⟩$ = 2, 4, 6, 8, 10, 12, 14.   

* For each value, generate ten different configurations of the benchmark and apply the Louvain algorithm on each of them. 
* Use the normalized mutual information to compute the similarity of each partition to the planted partition of the benchmark.   
* Calculate the average and the standard deviation of the similarity over the ten realizations and plot the average as a function of $⟨k^{ext}⟩$, using the standard deviation for error bars. 
* Then do the previous three steps again for asynchronous Label propapation. 


**Compare the findings of Louvain and Label propagation. What do you observe? Why?**

```python
 # apply Louvain
kext_list = [2, 4, 6, 8, 10, 12, 14]

# create a list to store mean and stdev of the score
score_mean_kext_list_1 = []
score_stdev_kext_list_1 = []

# for each kext
for kext in kext_list:
    
    # generate benchmark for each kext
    GN, q, nc, p_int, p_ext = GN_benchmark(kext)

    # created planted partition for each benchmark
    GPP = nx.generators.planted_partition_graph(q, nc, p_int, p_ext)
    
    # extract partition map
    planted_partition_map = nx.get_node_attributes(GPP,'block')
    
    # creat a score list for each kext
    score_list = []
    
    # generate 10 configurations for each benchmark GN
    for i in range(10):
        
        # generate configuration of the benchmark
        GC = nx.configuration_model(dict(GN.degree).values())

        # then apply louvain algorithm on each config
        louvain_partition_map = community_louvain.best_partition(GC)

        # use the normalized mutual information to compute the similarity
        labels_pred = list(dict(louvain_partition_map).values())
        labels_true = list(dict(planted_partition_map).values())
        score = normalized_mutual_info_score(labels_true,labels_pred)
        score_list.append(score)
    
    # get the mean and stdev of similairty for each kext
    score_mean = statistics.mean(score_list)
    score_stdev = statistics.stdev(score_list)
    
    # get the list of mean and stdev of similairty for each kext
    score_mean_kext_list_1.append(score_mean)
    score_stdev_kext_list_1.append(score_stdev)
```

```python
 # apply Label propagation (asyn)
kext_list = [2, 4, 6, 8, 10, 12, 14]

# create a list to store mean and stdev of the score
score_mean_kext_list_2 = []
score_stdev_kext_list_2 = []

# for each kext
for kext in kext_list:
    
    # generate benchmark for each kext
    GN, q, nc, p_int, p_ext = GN_benchmark(kext)

    # created planted partition for each benchmark
    GPP = nx.generators.planted_partition_graph(q, nc, p_int, p_ext)
    
    # extract partition map
    planted_partition_map = nx.get_node_attributes(GPP,'block')
    
    # creat a score list for each kext
    score_list = []
    
    # generate 10 configurations for each benchmark GN
    for i in range(10):
        
        # generate configuration of the benchmark
        GC = nx.configuration_model(dict(GN.degree).values())

        # then apply Label propagation (asyn) on each config
        label_partition = community.asyn_lpa_communities(GC)

        # convert parition into partition map
        label_partition_map = {}
        for idx, cluster_nodes in enumerate(list(label_partition)):
            for node in cluster_nodes:
                label_partition_map[node] = idx
            
        # use the normalized mutual information to compute the similarity
        labels_pred = list(dict(label_partition_map).values())
        labels_true = list(dict(planted_partition_map).values())
        score = normalized_mutual_info_score(labels_true,labels_pred)
        score_list.append(score)
    
    # get the mean and stdev of similairty for each kext
    score_mean = statistics.mean(score_list)
    score_stdev = statistics.stdev(score_list)
    
    # get the list of mean and stdev of similairty for each kext
    score_mean_kext_list_2.append(score_mean)
    score_stdev_kext_list_2.append(score_stdev)
```

```python
# report the average and the standard deviation in the plot
x = kext_list
y1 = score_mean_kext_list_1
yerr1 = score_stdev_kext_list_1
plt.errorbar(x, y1, yerr=yerr1, label = "Louvain")

y2 = score_mean_kext_list_2
yerr2 = score_stdev_kext_list_2
plt.errorbar(x, y2, yerr=yerr2,  label = "Label Propagation(asyn)")
plt.xlabel("Expected External Degree (kext)")
plt.ylabel("Average Similairty")
plt.title("Similarity between Louvain to Planted Partition vs. Label Propagation to Planted Partition")
leg = plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left');
```

From the graph above, we can see that there is little similairty between between Louvain and Planted Partition, and it is around 0.05 on average. There is no similarity between Label Propagation and Planted Partition as it the NMI value is zero.  NMI has an expected value of zero when two random partitions are compared, and both label Propagation and GN benchmark networks are random partitions per their definiation.


## Q3. Communities vs Erd ̋os-R ́enyi random networks
Create 10 different **Erd ̋os-R ́enyi random networks** with N = 1000 nodes and L = 5000, 10,000, 15,000, 20,000, 25,000, and 30,000 links.   

**Apply the Louvain algorithm to each network.** 
* Check the modularity values of the resulting partitions: are they close to zero? 
* Plot the average number of communities as a function of the average degree of the random networks: what conclusion does the trend suggest?
* Plot the average modularity of the partitions found for each value of L as a function of L (with standard deviation to indicate the error).  

**Label propagation (asyn)**
* Determine the average number of clusters as a function of L with label propagation (asyn) and compare the curve with the one from Louvain.

```python
# Apply the Louvain algorithm

L_list = [5000, 10000, 15000, 20000, 25000, 30000]

mean_modularity_L_list = []
stdev_modularity_L_list = []
mean_communities_L_list = []
mean_mean_degree_L_list = []

# create a function converts partition map into partition
def partition_convert(partition_map):
    partition= {}
    for node, p in partition_map.items():
        if p not in partition:
            partition[p] = [node]
        else:
            partition[p].append(node)
    partition = list(partition.values())
    return partition

# create a Erd ̋os-R ́enyi random network for each L
for L in L_list:   
    
    # create modularity list for each L
    modularity_L_list = []
  
    # create number of communities list for each L
    communities_L_list = []
    
    # create degree of graph for each L
    mean_deg_L_list = []
    
    # generate 10 different networks
    for i in range(10):
        G = nx.gnm_random_graph(1000, L)
        
        # Apply the Louvain algorithm
        partition_map = community_louvain.best_partition(G)
        partition = partition_convert(partition_map)
        
        # generate modularity for each L and append to a list
        modularity = nx.community.quality.modularity(G, partition)
        modularity_L_list.append(modularity)
        
        # get the number of communities and append to a list
        number_of_communities = len(partition)
        communities_L_list.append(number_of_communities)
        
        # get the average degree and appeend to a list
        degree_sequence = [G.degree(n) for n in G.nodes]
        mean_deg = statistics.mean(degree_sequence)
        mean_deg_L_list.append(mean_deg)

    # get the average number of communities for each L and append to a list
    mean_mean_degree_L = statistics.mean(mean_deg_L_list)
    mean_mean_degree_L_list.append(mean_mean_degree_L)

    # get the average degree oof the graph for each L and append to a list
    mean_communities_L = statistics.mean(communities_L_list)
    mean_communities_L_list.append(mean_communities_L)
    
    # get the average modularity for each L and append to a list
    mean_modularity_L = statistics.mean(modularity_L_list)
    mean_modularity_L_list.append(mean_modularity_L)
    
    # get the stdev modularity for each L and append to a list
    stdev_modularity_L = statistics.stdev(modularity_L_list)
    stdev_modularity_L_list.append(stdev_modularity_L)
```

```python
print("Average modularity of the resulting partition", statistics.mean(mean_modularity_L_list))
```

 Per above, we can see that the average modularity values of the resulting partitions are close to 0.17

```python
# Plot the average number of communities as a function of the average degree of the random networks:
x = mean_mean_degree_L_list
y = mean_communities_L_list

plt.plot(x, y) 
plt.xlabel('Average degree of the random networks')
plt.ylabel('Average number of communities')
plt.title('Erd ̋os-R ́enyi random networks under Louvain');
```

Per graph above, we can see that as average degree increases, the average number of communities decreases sharply till the average degree hits 20. After that, the average number of communities stabalize around 10 when average degree of random network increases. The trend indicates that as the average degree of the network increases, the size of comminites becomes stablized.

```python
# Plot the average modularity of the partitions found for each value of L as a function of L 
# with standard deviation to indicate the error

x = L_list
y = mean_modularity_L_list
yerr = stdev_modularity_L_list

plt.errorbar(x, y, yerr=yerr)
plt.xlabel('Number of Links')
plt.ylabel('Average Modularity')
plt.title('Average Modularity of Erd ̋os-R ́enyi random networks under Louvain');
```

```python
# Apply Label propagation (asyn)

L_list = [5000, 10000, 15000, 20000, 25000, 30000]

mean_modularity_L_list_2 = []
mean_communities_L_list_2 = []
mean_mean_degree_L_list_2 = []
stdev_modularity_L_list_2 = []

# create a Erd ̋os-R ́enyi random network for each L
for L in L_list:   

    # create modularity list for each L
    modularity_L_list = []
    # create number of communities list for each L
    communities_L_list = []
    # create degree of graph for each L
    mean_deg_L_list = []
    
    # generate 10 different networks
    for i in range(10):
        G = nx.gnm_random_graph(1000, L)
        
        # Apply Label propagation (asyn)
        partition = list(community.asyn_lpa_communities(G))
        
        # generate modularity for each L and append to a list
        modularity = nx.community.quality.modularity(G, partition)
        modularity_L_list.append(modularity)
        
        # get the number of communities and append to a list
        number_of_communities = len(partition)
        communities_L_list.append(number_of_communities)
        
        # get the average degree and appeend to a list
        degree_sequence = [G.degree(n) for n in G.nodes]
        mean_deg = statistics.mean(degree_sequence)
        mean_deg_L_list.append(mean_deg)

    # get the average number of communities for each L and append to a list
    mean_mean_degree_L = statistics.mean(mean_deg_L_list)
    mean_mean_degree_L_list.append(mean_mean_degree_L)

    
    # get the average degree oof the graph for each L and append to a list
    mean_communities_L = statistics.mean(communities_L_list)
    mean_communities_L_list_2.append(mean_communities_L)
    
    # get the average modularity for each L and append to a list
    mean_modularity_L = statistics.mean(modularity_L_list)
    mean_modularity_L_list_2.append(mean_modularity_L)
    
    # get the stdev modularity for each L and append to a list
    stdev_modularity_L = statistics.stdev(modularity_L_list)
    stdev_modularity_L_list_2.append(stdev_modularity_L)
```

## Q4. Communities vs Barabasi-Albert networks
Create 10 different **Barabasi-Albert networks** with N=1000 and m =2, 5, 10, 15, 20, 50.
